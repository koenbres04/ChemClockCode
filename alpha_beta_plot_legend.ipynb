```python
import matplotlib.patches as mpatches
fig, ax = plt.subplots(1, figsize=(7,7))
ax.imshow(plt.imread("alpha_beta_plot_no_legend.png"), extent=[0, 6.26, 0, 5.08])
ax.set_xlabel(r"$\alpha$")
ax.set_ylabel(r"$\beta$")
legend_elements = [mpatches.Patch(color=(250/255, 60/255, 95/255), label="unstable node"),
                  mpatches.Patch(color=(75/255, 80/255, 210/255), label="unstable focus"),
                  mpatches.Patch(color=(125/255, 125/255, 125/255), label="stable focus"),
                  mpatches.Patch(color=(250/255, 175/255, 60/255), label="stable node"),
                  mpatches.Patch(color=(0/255, 0/255, 0/255), label="centre")]
ax.legend(handles=legend_elements, fontsize="8")
plt.savefig("alpha_beta_plot.png")
```
